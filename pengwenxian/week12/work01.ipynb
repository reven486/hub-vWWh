## 数据库解析

```python
'''数据库解析'''
from typing import Union
import traceback
from sqlalchemy import create_engine, inspect, func, select, Table, MetaData, text # ORM 框架
import pandas as pd

class DBParser:
    '''DBParser 数据库的解析'''
    def __init__(self, db_url:str) -> None:
        '''初始化
        db_url: 数据库链接地址
        mysql: mysql://root:123456@localhost:3306/mydb?charset=utf8mb4
        sqlite: sqlite://chinook.db
        '''

        # 判断数据库类型
        if 'sqlite' in db_url:
            self.db_type = 'sqlite'
        elif 'mysql' in db_url:
            self.db_type = 'mysql'

        # 链接数据库
        self.engine = create_engine(db_url, echo=False)
        self.conn = self.engine.connect()
        self.cursor = self.conn.cursor()
        self.db_url = db_url

        # 查看表明
        self.inspector = inspect(self.engine)
        self.table_names = self.inspector.get_table_names() # 获取table信息

        self._table_fields = {} # 数据表字段
        self.foreign_keys = [] # 数据库外键
        self._table_sample = {} # 数据表样例

        # 依次对每张表的字段进行统计
        for table_name in self.table_names:
            print("Table ->", table_name)
            self._table_fields[table_name] = {}

            # 累计外键
            self.foreign_keys += [
                {
                    'constrained_table': table_name,
                    'constrained_columns': x['constrained_columns'],
                    'referred_table': x['referred_table'],
                    'referred_columns': x['referred_columns'],
                } for x in self.inspector.get_foreign_keys(table_name)
            ]

            # 获取当前表的字段信息
            table_instance = Table(table_name, MetaData(), autoload_with=self.engine)
            table_columns = self.inspector.get_columns(table_name)
            self._table_fields[table_name] = {x['name']:x for x in table_columns}

            # 对当前字段进行统计
            for column_meta in table_columns:
                # 获取当前字段
                column_instance = getattr(table_instance.columns, column_meta['name'])

                # 统计unique
                query = select(func.count(func.distinct(column_instance)))
                distinct_count = self.conn.execute(query).fetchone()[0]
                self._table_fields[table_name][column_meta['name']]['distinct'] = distinct_count

                # 统计most frequency value
                field_type = self._table_fields[table_name][column_meta['name']]['type']
                field_type = str(field_type)
                if 'text' in field_type.lower() or 'char' in field_type.lower():
                    query = (
                        select(column_instance, func.count().label('count'))
                        .group_by(column_instance)
                        .order_by(func.count().desc())
                        .limit(1)
                    )
                    top1_value = self.conn.execute(query).fetchone()[0]
                    self._table_fields[table_name][column_meta['name']]['mode'] = top1_value

                # 统计missing个数
                query = select(func.count()).filter(column_instance == None)
                nan_count = self.conn.execute(query).fetchone()[0]
                self._table_fields[table_name][column_meta['name']]['nan_count'] = nan_count

                # 统计max
                query = select(func.max(column_instance))
                max_value = self.conn.execute(query).fetchone()[0]
                self._table_fields[table_name][column_meta['name']]['max'] = max_value

                # 统计min
                query = select(func.min(column_instance))
                min_value = self.conn.execute(query).fetchone()[0]
                self._table_fields[table_name][column_meta['name']]['min'] = min_value

                # 任意取值
                query = select(column_instance).limit(10)
                random_value = self.conn.execute(query).all()
                random_value = [x[0] for x in random_value]
                random_value = [str(x) for x in random_value if x is not None]
                random_value = list(set(random_value))
                self._table_fields[table_name][column_meta['name']]['random'] = random_value[:3]

            # 获取表样例（第一行）
            query = select(table_instance)
            self._table_sample[table_name] = pd.DataFrame([self.conn.execute(query).fetchone()])
            self._table_sample[table_name].columns = [x['name'] for x in table_columns]

    def get_table_fields(self, table_name) -> pd.DataFrame:
        '''获取表字段信息'''
        return pd.DataFrame.from_dict(self._table_fields[table_name]).T

    def get_table_num(self) -> int:
        '''获取数据库表数量'''
        return len(self.table_names)

    def get_table_rows(self, table_name) -> int:
        '''获取数据表行数'''
        sql = f"SELECT COUNT(*) FROM {table_name}"
        row_count = self.cursor.execute(sql).fetchone()[0]
        return row_count

    def get_data_relations(self) -> pd.DataFrame:
        '''获取数据库链接信息（主键和外键）'''
        return pd.DataFrame(self.foreign_keys)

    def get_table_sample(self, table_name) -> pd.DataFrame:
        '''获取数据表样例'''
        return self._table_sample[table_name]

    def check_sql(self, sql) -> Union[bool, str]:
        '''检查sql是否合理

        参数
            sql: 待执行句子

        返回: 是否可以运行 报错信息
        '''
        try:
            self.conn.execute(text(sql))
            return True, 'ok'
        except:
            err_msg = traceback.format_exc()
            return False, err_msg

    def execute_sql(self, sql) -> bool:
        '''运行SQL'''
        result = self.conn.execute(text(sql))
        return list(result)
```

## Chinook 问答 Agent
基于 `chinook.db` 的一个轻量级 NL2SQL Agent，优先使用规则覆盖高频统计问题，并预留 LLM 扩展能力。

```python
import re
import sqlite3
import pandas as pd

class ChinookNL2SQLAgent:
    def __init__(self, db_path='./chinook.db', parser=None, llm_func=None):
        self.db_path = db_path
        self.parser = parser
        self.llm_func = llm_func

    def _normalize_question(self, question: str) -> str:
        return re.sub(r'\s+', '', question.lower())

    def _build_schema_prompt(self, question: str) -> str:
        if self.parser is None:
            raise ValueError('未提供 parser，无法构造 schema prompt')

        schema_lines = []
        for table_name in self.parser.table_names:
            columns = list(self.parser.get_table_fields(table_name).index)
            schema_lines.append(f'- {table_name}: {', '.join(columns)}')

        return f'''你是一个专业的 SQLite SQL 专家。请根据数据库 schema 将用户问题翻译为可执行 SQL。
只输出 SQL，不要输出解释，不要加 markdown 代码块。

数据库 schema:
{chr(10).join(schema_lines)}

用户问题: {question}
'''

    def _rule_nl2sql(self, question: str):
        q = self._normalize_question(question)

        if ('数据库' in q and ('多少张表' in q or '几张表' in q)):
            return "SELECT COUNT(*) AS table_count FROM sqlite_master WHERE type='table' AND name NOT LIKE 'sqlite_%';"

        if (('员工表' in q or 'employees' in q) and ('多少条记录' in q or '多少条' in q or '记录数' in q or '多少行' in q)):
            return "SELECT COUNT(*) AS employee_count FROM employees;"

        if ('客户' in q and '员工' in q and ('分别' in q or '各' in q)):
            return "SELECT (SELECT COUNT(*) FROM customers) AS customer_count, (SELECT COUNT(*) FROM employees) AS employee_count;"

        return None

    def nl2sql(self, question: str) -> str:
        sql = self._rule_nl2sql(question)
        if sql is not None:
            return sql

        if self.llm_func is None:
            raise ValueError('当前问题未命中内置规则，且未配置 llm_func')

        prompt = self._build_schema_prompt(question)
        sql = self.llm_func(prompt).strip()
        sql = sql.strip('`').replace('sql\n', '').strip()
        return sql

    def execute_sql(self, sql: str):
        with sqlite3.connect(self.db_path) as conn:
            cursor = conn.cursor()
            cursor.execute(sql)
            rows = cursor.fetchall()
            columns = [x[0] for x in cursor.description] if cursor.description else []
        return columns, rows

    def _format_answer(self, question: str, columns, rows) -> str:
        q = self._normalize_question(question)
        if not rows:
            return '查询成功，但结果为空。'

        first_row = rows[0]
        if '多少张表' in q or '几张表' in q:
            return f'数据库中总共有 {first_row[0]} 张表。'

        if '员工表' in q and ('多少条记录' in q or '多少条' in q or '记录数' in q or '多少行' in q):
            return f'员工表中共有 {first_row[0]} 条记录。'

        if '客户' in q and '员工' in q and ('分别' in q or '各' in q):
            customer_count = first_row[0]
            employee_count = first_row[1]
            return f'数据库中客户共有 {customer_count} 个，员工共有 {employee_count} 个。'

        if len(rows) == 1 and len(first_row) == 1:
            return f'查询结果为 {first_row[0]}。'

        return pd.DataFrame(rows, columns=columns).to_markdown(index=False)

    def answer(self, question: str):
        sql = self.nl2sql(question)
        columns, rows = self.execute_sql(sql)
        answer = self._format_answer(question, columns, rows)
        return {
            'question': question,
            'sql': sql,
            'columns': columns,
            'rows': rows,
            'answer': answer,
        }
```

```python
agent = ChinookNL2SQLAgent(db_path='./chinook.db', parser=parser, llm_func=ask_qwen)

demo_questions = [
    '数据库中总共有多少张表？',
    '员工表中有多少条记录？',
    '在数据库中所有客户个数和员工个数分别是多少？',
]

demo_results = []
for question in demo_questions:
    result = agent.answer(question)
    demo_results.append({
        'question': result['question'],
        'sql': result['sql'],
        'answer': result['answer'],
    })

pd.DataFrame(demo_results)
```
